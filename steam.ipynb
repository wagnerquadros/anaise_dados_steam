## 4. Limpeza e Pré-processamento dos Dados

Nesta etapa, os dados são preparados para análise, garantindo consistência e confiabilidade nas métricas calculadas.

As principais ações realizadas são:

- Conversão da variável `owners` de texto para valor numérico
- Ajuste dos valores de preço (centavos → dólares)
- Filtragem de jogos válidos para análise:
  - apenas jogos pagos
  - jogos com tempo de jogo maior que zero
  - jogos com pelo menos uma avaliação

### 4.1 Tratamento da variável `owners`

A variável `owners` é apresentada no dataset como um intervalo de valores, por exemplo:

"100,000 .. 200,000"

Esse formato textual não é adequado para análises quantitativas. Por esse motivo, foi realizada uma transformação para obter um valor numérico representativo.

A estratégia adotada foi calcular o ponto médio do intervalo:

(100000 + 200000) / 2 = 150000

Esse valor passa a representar uma estimativa da quantidade de proprietários do jogo, sendo utilizado como uma proxy de popularidade nas análises subsequentes.

```python
# Função para converter intervalo em ponto médio
def tratar_owners(valor):
    partes = valor.replace(",", "").split("..")
    minimo = float(partes[0].strip())
    maximo = float(partes[1].strip())
    return (minimo + maximo) / 2

# Aplicar no dataframe
df["owners_mid"] = df["owners"].apply(tratar_owners)

# Verificar
df[["owners", "owners_mid"]].head()
```

### 4.2 Converter preços

```python
df["price_usd"] = df["price"] / 100
df["initialprice_usd"] = df["initialprice"] / 100

df[["price", "price_usd"]].head()
```

### 4.3 Filtrar jogos pagos

```python
df_pagos = df[df["price"] > 0].copy()

print("Total original:", len(df))
print("Jogos pagos:", len(df_pagos))
```

### 4.4 Filtrar jogos com tempo válido

```python
df_validos = df_pagos[df_pagos["average_forever"] > 0].copy()

print("Após filtro de tempo:", len(df_validos))
```

### 4.5 Filtrar jogos com avaliação

```python
total_reviews = df_validos["positive"] + df_validos["negative"]

df_validos = df_validos[total_reviews > 0].copy()

print("Após filtro de avaliações:", len(df_validos))
```
